```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#Extract CSV INTO DATA FRAME
World_population ="Resources/World_Population.csv"
```

```python
World_sucide ="Resources/World_SucideRate.csv"
```

```python
population_df = pd.read_csv(World_population)
sucide_df = pd.read_csv(World_sucide)
```

```python
population_df.head()
```

```python
#Renaming un-named column to country 
population_df = population_df.rename(columns={'Unnamed: 0': 'country'})
#population_df['country_code'] = ['N/A'] * len(population_df['country'])
population_df
```

```python
# Taking first 3 character from the country column to new column country code colmun 
population_df['country_code'] = ['N/A'] * len(population_df['country'])

for i in range(len(population_df['country_code'])):
    code = population_df.loc[i, 'country']
    population_df.loc[i, 'country_code'] = code[:3]
```

```python
population_df
```

```python
cols = ['country','country_code','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']
new_column = population_df[cols]
rename_col_dict = {
    "country": "countryName",
    "country_code": "country_code",
    "2000" : "year_2000",
    "2001" : "year_2001", 
    "2002" : "year_2002",
    "2003": "year_2003", 
    "2004" : "year_2004", 
    "2005": "year_2005", 
    "2006": "year_2006", 
    "2007": "year_2007", 
    "2008": "year_2008", 
    "2009": "year_2009", 
    "2010" : "year_2010"
}

new_column = new_column.rename(columns =rename_col_dict)
new_column = new_column.drop_duplicates("country_code")
new_column = new_column.dropna()
new_column = new_column.set_index("country_code")
new_column.head()
```
